```python
def clean_makerting_customer_analysis (x):
    x = x.rename(columns={'EmploymentStatus':'Employment Status'})
    x['Effective To Date'] = x['Effective To Date'].astype('datetime64[ns]')
    x = x.dropna()
            
    return x    
```

```python
data = clean_makerting_customer_analysis(data)
data.head(1)
```
